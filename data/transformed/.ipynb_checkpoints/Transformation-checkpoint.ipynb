# merge dataframes

```python
dfdb = dfdb.drop("row_num", axis=1)
```

```python
dfapi = dfapi.rename(columns={"Payment": "Payement"})
```

```python
dfjson = dfjson.rename(columns={"Payment": "Payement"})
```

```python
dfapi = dfjson.rename(columns={"Shipping Costs": "Shipping_Costs"})
```

```python
dfjson = dfjson.rename(columns={"Shipping Costs": "Shipping_Costs"})
```

```python
df = pd.concat([dfapi, dfdb, dfjson], ignore_index=True)
```

```python
# Valeur absolue dans les colonnes Age et Quantity
df['Age'] = df['Age'].abs()
df['Quantity'] = df['Quantity'].abs()
```

```python
# Remplacer les valeurs nulles dans la colonne rating_customer par 0
df['rating_customer'].fillna(0, inplace=True)
```

```python
# Supprimer les lignes où la valeur dans la colonne Quantity est égale à 0
df = df[df['Quantity'] != 0]
```

```python
# import pandas as pd
from cryptography.fernet import Fernet
import base64
from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.primitives.kdf.pbkdf2 import PBKDF2HMAC

# Encrypt a value using the key
def encrypt_value(value, key):
    encrypted_value = key.encrypt(value.encode())
    return encrypted_value.decode()

# Encrypt the columns
def encrypt_columns(df, columns, key):
    for column in columns:
        df[column] = df[column].apply(lambda x: encrypt_value(str(x), key))
    return df

# Decrypt a value using the key
def decrypt_value(value, key):
    decrypted_value = key.decrypt(value.encode())
    return decrypted_value.decode()

# Decrypt the columns
def decrypt_columns(df, columns, key):
    for column in columns:
        df[column] = df[column].apply(lambda x: decrypt_value(x, key))
    return df

# Example usage
password = "123@123"
salt = b'\x07\xd1\x85\xeb\x1b\xad`a\xd5\xdd\x11\x11\x9c\x0e\xb0'

# Derive the key from the password and salt
kdf = PBKDF2HMAC(
    algorithm=hashes.SHA256(),
    length=32,
    salt=salt,
    iterations=100000
)
key = base64.urlsafe_b64encode(kdf.derive(password.encode()))

# Create an instance of the Fernet cipher using the key
cipher_suite = Fernet(key)

# Encrypt the columns
columns_to_encrypt = ['Customer_Name','Email', 'Phone', 'Adresse']
encrypted_df = encrypt_columns(df, columns_to_encrypt, cipher_suite)
```

# Loading datafram into my database in sql server

```python
import pandas as pd
import pyodbc

# Define your database connection parameters
server = 'localhost'
database = 'PFR_Data'
username = 'yassine'
password = '123@123'

# Create a connection string
conn_str = f'DRIVER={{SQL Server}};SERVER={server};DATABASE={database};UID={username};PWD={password};'

# Connect to the database
conn = pyodbc.connect(conn_str)

# Create a cursor from the connection
cursor = conn.cursor()
```

```python
# Insert data into Dim_Customer
cursor.executemany('INSERT INTO Dim_Customer (Customer_Name, Age, Gender, Location, Customer_Segment, Email, Phone, Adresse, Rating, Payement) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)',
                   df[['Customer_Name', 'Age', 'Gender', 'Location', 'Customer_Segment', 'Email', 'Phone', 'Adresse', 'rating_customer', 'Payement']].values.tolist())

# Insert data into Dim_Product
cursor.executemany('INSERT INTO Dim_Product (Product, Category, Subcategory) VALUES (?, ?, ?)',
                   df[['Product', 'Category', 'Subcategory']].values.tolist())

# Insert data into Dim_Shipping
cursor.executemany('INSERT INTO Dim_Shipping (Shipping_Method, Shipper, Shipping_Costs) VALUES (?, ?, ?)',
                   df[['Shipping_Method', 'Shipper', 'Shipping_Costs']].values.tolist())

# Insert data into Dim_Date
cursor.executemany('INSERT INTO Dim_Date (Date) VALUES (?)',
                   df[['Date_Order']].values.tolist())

# Insert data into Dim_Warehouse
cursor.executemany('INSERT INTO Dim_Warehouse (Warhouse_Name, Location_warehouse) VALUES (?, ?)',
                   df[['Warehouse_Name', 'Location_warehouse']].values.tolist())

# Insert data into Dim_Supplier
cursor.executemany('INSERT INTO Dim_Supplier (Location_Supplier, Supplier_Name, Supplier_Phone, Quality_Score) VALUES (?, ?, ?, ?)',
                   df[['Location_Supplier', 'Supplier_Name', 'Supplier_Phone', 'Quality_Score']].values.tolist())
```

```python
# Insert data into Fact_Sales
cursor.executemany('INSERT INTO Fact_Sales (Id_Date_Order, Id_Product, Id_Shipping, Price, Quantity, Weight, Tax_5_percent) VALUES (?, ?, ?, ?, ?, ?, ?)',
                   df[['Date_Order', 'Product', 'Shipping_Method', 'Price', 'Quantity', 'Weight', 'Tax_5_percent']].values.tolist())

# Insert data into Fact_Inventory
cursor.executemany('INSERT INTO Fact_Inventory (Id_Supplier, Id_Restock_Date, Id_Product, Id_Warehouse, StockLevel, Restock_Threshold) VALUES (?, ?, ?, ?, ?, ?)',
                   df[['Supplier_Name', 'Date_Order', 'Product', 'Warehouse_Name', 'StockLevel', 'Restock_Threshold']].values.tolist())
```

```python
# Commit the changes and close the connection
conn.commit()
conn.close()
```
