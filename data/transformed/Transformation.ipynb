```python
# Upsert data into Dim_Customer
cursor.execute("""
    MERGE INTO Dim_Customer AS target
    USING (
        SELECT DISTINCT
            Customer_Name,
            Age,
            Gender,
            Location,
            Customer_Segment,
            Email,
            Phone,
            Adresse,
            Payement
        FROM Dim_Customer
    ) AS source
    ON target.Customer_Name = source.Customer_Name
    WHEN NOT MATCHED THEN
        INSERT (
            Customer_Name,
            Age,
            Gender,
            Location,
            Customer_Segment,
            Email,
            Phone,
            Adresse,
            Payement
        ) VALUES (
            source.Customer_Name,
            source.Age,
            source.Gender,
            source.Location,
            source.Customer_Segment,
            source.Email,
            source.Phone,
            source.Adresse,
            source.Payement
        );
""")

# Upsert data into Dim_Product
cursor.execute("""
    MERGE INTO Dim_Product AS target
    USING (
        SELECT DISTINCT
            Product,
            Category,
            Subcategory
        FROM Dim_Product
    ) AS source
    ON target.Product = source.Product
    WHEN NOT MATCHED THEN
        INSERT (
            Product,
            Category,
            Subcategory
        ) VALUES (
            source.Product,
            source.Category,
            source.Subcategory
        );
""")

# Upsert data into Dim_Supplier
cursor.execute("""
    MERGE INTO Dim_Supplier AS target
    USING (
        SELECT DISTINCT
            Location_Supplier,
            Supplier_Name,
            Supplier_Phone,
            Quality_Score
        FROM Dim_Supplier
    ) AS source
    ON target.Supplier_Name = source.Supplier_Name
    WHEN NOT MATCHED THEN
        INSERT (
            Location_Supplier,
            Supplier_Name,
            Supplier_Phone,
            Quality_Score
        ) VALUES (
            source.Location_Supplier,
            source.Supplier_Name,
            source.Supplier_Phone,
            source.Quality_Score
        );
""")

# Upsert data into Dim_Warehouse
cursor.execute("""
    MERGE INTO Dim_Warehouse AS target
    USING (
        SELECT DISTINCT
            Warhouse_Name,
            Location_warehouse
        FROM Dim_Warehouse
    ) AS source
    ON target.Warhouse_Name = source.Warhouse_Name
    WHEN NOT MATCHED THEN
        INSERT (
            Warhouse_Name,
            Location_warehouse
        ) VALUES (
            source.Warhouse_Name,
            source.Location_warehouse
        );
""")
```

```python
# Upsert data into Fact_Sales
cursor.execute("""
    MERGE INTO Fact_Sales AS target
    USING (
        SELECT 
            c.Id_Customer,
            d.Id_Date,
            p.Id_Product,
            s.Id_Shipping,
            MAX(f.Price) AS Price,
            MAX(f.Quantity) AS Quantity,
            MAX(f.Weight) AS Weight,
            MAX(f.Tax_5_percent) AS Tax_5_percent,
            MAX(f.Shipping_Costs) AS Shipping_Costs,
            MAX(f.Rating) AS Rating
        FROM Fact_Sales f
            JOIN Dim_Customer c ON f.Id_Customer = c.Id_Customer
            JOIN Dim_Date d ON f.Id_Date_Order = d.Id_Date
            JOIN Dim_Product p ON f.Id_Product = p.Id_Product
            JOIN Dim_Shipping s ON f.Id_Shipping = s.Id_Shipping
        GROUP BY
            c.Id_Customer,
            d.Id_Date,
            p.Id_Product,
            s.Id_Shipping
    ) AS source
    ON target.Id_Customer = source.Id_Customer
    AND target.Id_Date_Order = source.Id_Date
    AND target.Id_Product = source.Id_Product
    AND target.Id_Shipping = source.Id_Shipping
    WHEN MATCHED THEN
        UPDATE SET
            Price = source.Price,
            Quantity = source.Quantity,
            Weight = source.Weight,
            Tax_5_percent = source.Tax_5_percent,
            Shipping_Costs = source.Shipping_Costs,
            Rating = source.Rating
    WHEN NOT MATCHED THEN
        INSERT (
            Id_Customer,
            Id_Date_Order,
            Id_Product,
            Id_Shipping,
            Price,
            Quantity,
            Weight,
            Tax_5_percent,
            Shipping_Costs,
            Rating
        ) VALUES (
            source.Id_Customer,
            source.Id_Date,
            source.Id_Product,
            source.Id_Shipping,
            source.Price,
            source.Quantity,
            source.Weight,
            source.Tax_5_percent,
            source.Shipping_Costs,
            source.Rating
        );
""")
```

```python
# Upsert data into Fact_Inventory
cursor.execute("""
    MERGE INTO Fact_Inventory AS target
    USING (
        SELECT 
            s.Id_Supplier,
            d.Id_Date,
            p.Id_Product,
            w.Id_Warehouse,
            f.Inventory_id,
            f.StockLevel,
            f.Restock_Threshold
        FROM
            Fact_Inventory f
            JOIN Dim_Supplier s ON f.Id_Supplier = s.Id_Supplier
            JOIN Dim_Date d ON f.Id_Restock_Date = d.Id_Date
            JOIN Dim_Product p ON f.Id_Product = p.Id_Product
            JOIN Dim_Warehouse w ON f.Id_Warehouse = w.Id_Warehouse
    ) AS source
    ON target.Id_Supplier = source.Id_Supplier
    AND target.Id_Restock_Date = source.Id_Date
    AND target.Id_Product = source.Id_Product
    AND target.Id_Warehouse = source.Id_Warehouse
    WHEN NOT MATCHED THEN
        INSERT (
            Id_Supplier,
            Id_Restock_Date,
            Id_Product,
            Id_Warehouse,
            Inventory_id,
            StockLevel,
            Restock_Threshold
        )
        VALUES (
            source.Id_Supplier,
            source.Id_Date,
            source.Id_Product,
            source.Id_Warehouse,
            source.Inventory_id,
            source.StockLevel,
            source.Restock_Threshold
        )
    WHEN MATCHED THEN
        UPDATE SET
            StockLevel = source.StockLevel,
            Restock_Threshold = source.Restock_Threshold;
""")
```
